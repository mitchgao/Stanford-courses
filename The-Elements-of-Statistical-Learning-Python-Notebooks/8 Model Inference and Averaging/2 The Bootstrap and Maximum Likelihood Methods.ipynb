```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
t = np.array([0., 0., 0., 0., 0.7, 1.4, 2.2, 3., 3., 3., 3.])
```

```python
def B(x, k, i, t):
    if k == 0:
        return 1.0 if t[i] <= x < t[i+1] else 0.0
    if t[i+k] == t[i]:
        c1 = 0.0
    else:
        c1 = (x - t[i])/(t[i+k] - t[i]) * B(x, k-1, i, t)
    if t[i+k+1] == t[i+1]:
        c2 = 0.0
    else:
        c2 = (t[i+k+1] - x)/(t[i+k+1] - t[i+1]) * B(x, k-1, i+1, t)
    return c1 + c2
```

```python
x = np.linspace(0, 3, 400)[:-1]
x_exp = np.zeros(shape=(x.shape[0], 7))
for n in range(7):
    for i in range(x.shape[0]):
        x_exp[i, n] = B(x[i], 3, n, t)
    plt.plot(x, x_exp[:,n])
```

```python
y = points[:,1]
def expand(points):
    X = np.zeros(shape=(points.shape[0], 7))
    for n in range(7):
        for i in range(X.shape[0]):
            X[i, n] = B(points[i, 0], 3, n, t)
    return X
X = expand(points)
```

```python
b = np.linalg.inv(X.T @ X) @ X.T @ y
y_hat = X @ b
y_smooth = x_exp @ b
```

```python
sigma_squared_hat = np.mean((y - y_hat)**2)
b_var_hat = np.linalg.inv(X.T @ X) * sigma_squared_hat
```

```python
se = np.zeros(shape=(x.shape[0]))
for i in range(x.shape[0]):
    se[i] = np.sqrt(x_exp[i] @ np.linalg.inv(X.T @ X) @ x_exp[i]) * np.sqrt(sigma_squared_hat)
```

```python
plt.scatter(points[:,0], points[:,1])
plt.plot(x, y_smooth)
plt.plot(x, y_smooth + 1.9 * se, '--', color='gray')
plt.plot(x, y_smooth - 1.9 * se, '--', color='gray')
```

```python
res = np.zeros(shape=(200, x.shape[0]))
plt.scatter(points[:,0], points[:,1])
for i in range(200):
    samples_idx = np.random.choice(list(range(50)), 50)
    X = expand(points[samples_idx, :])
    y = points[samples_idx, :][:,1]
    b = np.linalg.inv(X.T @ X) @ X.T @ y
    y_smooth_sample = x_exp @ b
    res[i, :] = y_smooth_sample
    plt.plot(x, y_smooth_sample, linewidth=0.3)
```

```python
low, hi = np.zeros(shape=x.shape[0]), np.zeros(shape=x.shape[0])
for i in range(x.shape[0]):
    a = np.argsort(res[:, i])
    low[i], hi[i] = res[a[5], i], res[a[-5], i]
```
